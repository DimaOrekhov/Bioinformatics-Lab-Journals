```python
def compose(folder, o_file):
    from os import listdir
    
    af=open(o_file, 'a')
    for file in listdir(folder):
        fid=open(folder+file, 'r')
        af.write(fid.read())
        fid.close()
```

```python
compose('Data/Human/', 'human.fasta')
```

```python
def multiple_folders(folders, o_file):
    from os import listdir
    
    af=open(o_file, 'a')
    
    for folder in folders:
        for file in listdir(folder):
            fid=open(folder+file, 'r')
            af.write(fid.read())
            fid.close()
```

```python
multiple_folders(['Data/Human/', 'Data/Neanderthal/', 'Data/Denisova/'], 'HSD.fasta')
```

```python
def get_pairwise_matrix(file):
    
    from tqdm import tqdm_notebook
    import pandas as pd
    import numpy as np
    from itertools import product
    
    def sim(ch1, ch2):
        return 0 if ch1==ch2 else 1
    
    fid=open(file)
    next(fid)
    next(fid)
    next(fid)
    
    st=fid.read()
    fid.close()
    st=st[:-1].split('\n\n')
    
    for i, part in tqdm_notebook(enumerate(st)):
        
        part=part.split('\n')
        cons=part[-1]
        part=[x.split('      ') for x in part if x!=part[-1]]
        offset=len(part[0][0])+len('      ')
        cons=list(cons[offset:])
        
        if i==0:
            idc=[x[0] for x in part]
            dist_mat=pd.DataFrame(np.zeros([len(idc), len(idc)], dtype=int), index=idc, columns=idc)
            
        temp=pd.DataFrame([list(x[1]) for x in part], index=idc)
        
        for j in (np.arange(0, len(cons))[np.array(cons)!='*']):
            for row, col in product(idc, idc):
                dist_mat.loc[row, col]+=sim(temp.loc[row, j], temp.loc[col, j])

    return dist_mat
```

```python
def seq_lengths(fasta):
    from Bio import SeqIO
    from numpy import mean
    
    lengths=[]
    for record in SeqIO.parse(fasta, 'fasta'):
        lengths.append(len(record.seq))
    return lengths, mean(lengths)
```
